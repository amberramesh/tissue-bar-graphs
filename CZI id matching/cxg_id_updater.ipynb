```python
import pandas as pd
import os
```

```python
#download link for S0 (old info): https://docs.google.com/spreadsheets/d/1cwxztPg9sLq0ASjJ5bntivUk6dSKHsVyR1bE6bXvMkY/edit#gid=1529271254

s0 = pd.read_csv('Table S0 and S1_ Experimental datasets for computing cell type populations  - Table S0_ Experimental datasets for computing cell type populations .csv')
s0
```

```python
s0_cxg = s0.loc[s0['source'] == 'CxG',['unique_dataset_id','non_hubmap_donor_id','organ','number_of_cells_total']].dropna()
s0_cxg
```

```python
s0_cxg['donor_organ'] = s0_cxg.apply(lambda row: "_".join([row['non_hubmap_donor_id'],row['organ']]), axis=1)
s0_cxg.drop(['non_hubmap_donor_id','organ'], axis=1, inplace=True)
s0_cxg
```

```python
s0_cxg['dataset_id'] = s0_cxg.apply(lambda row: row['unique_dataset_id'].split('_')[0], axis=1)
```

```python
#id_name_mapper gets dataset names for dataset IDs from CZI

id_name_mapper = pd.read_csv('id_name_mapper.csv', index_col=[0])
id_name_mapper = id_name_mapper.drop_duplicates()
```

```python
s0_cxg = s0_cxg.merge(id_name_mapper,how='left').drop('dataset_id',axis=1)
s0_cxg
```

```python
#creates a S0 like table from files currently in repo (new info)

info = []
for folder in os.listdir('csv'):
    if 'CxG' in folder:
        for csvfile in os.listdir(os.path.join('csv',folder)):
            data = pd.read_csv(os.path.join('csv',folder,csvfile))
            info.append([csvfile[:-4], data['count'].sum(), "_".join(csvfile.split("_")[1:])[:-4], data['dataset_name'].unique()[0]])

cxg_csvinfo = pd.DataFrame(info,columns=s0_cxg.columns)
cxg_csvinfo
```

```python
s0_cxg['number_of_cells_total'] = s0_cxg['number_of_cells_total'].astype(int)
```

```python
lookup = s0_cxg.merge(cxg_csvinfo,left_on=['donor_organ','number_of_cells_total','dataset_name'], right_on=['donor_organ','number_of_cells_total','dataset_name'], how='left')
lookup
```
